```python
# Import required libraries from Qiskit
from qiskit import QuantumCircuit
import numpy as np

# Define the QFT function
def quantum_fourier_transform(num_qubits):
    # Create a quantum circuit with the specified number of qubits
    circuit = QuantumCircuit(num_qubits)
    
    # Perform the Quantum Fourier Transform operations
    for qubit in range(num_qubits):
        # Apply a Hadamard gate to the current qubit
        circuit.h(qubit)
        # Apply controlled phase rotations to subsequent qubits
        for next_qubit in range(qubit + 1, num_qubits):
            phase_angle = np.pi / (2 ** (next_qubit - qubit))
            circuit.cp(phase_angle, qubit, next_qubit)
    
    # Reverse the order of qubits
    for i in range(num_qubits // 2):
        circuit.swap(i, num_qubits - i - 1)

    return circuit

# Specify the number of qubits for the QFT
num_qubits = 5
# Generate the QFT circuit
qft_circuit = quantum_fourier_transform(num_qubits)

# Visualize the QFT circuit
qft_circuit.draw('mpl')
```
